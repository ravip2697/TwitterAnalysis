```python
import pandas as pd

df = pd.read_csv('Twitter_Data.csv')
df = df.dropna()
df
```

```python
# Filter the tweets that contain motivational keywords
query = "motivation|inspire|success|positive|MotivationMonday|SuccessMindset|GratitudeAttitude|InspiringQuotes|Mindfulness|PersonalGrowth|SelfImprovement"
motivational_tweets = df[df['clean_text'].str.contains(query, case=False)]

# Extract the text of each tweet
motivational_tweets_text = motivational_tweets['clean_text'].tolist()

# Print the extracted motivational tweets
motivational_tweets_text
```

```python
clean_text_list = []

for text in motivational_tweets_text:
    clean_text = text.replace("\n", "")
    clean_text_list.append(clean_text)

print(clean_text_list)
```

```python
df = pd.DataFrame(clean_text_list, columns=["Tweet"])
df
```
